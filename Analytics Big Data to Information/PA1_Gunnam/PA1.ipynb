```python
GER = ['Porsche','Audi','BMW','Mercedes-Benz','Mini', 'Volkswagen']
JPN = ['Acura', 'Honda','Infiniti','Isuzu','Lexus','Mazda','Mitsubishi','Nissan','Scion','Subaru','Suzuki','Toyota']
KOR = ['Hyundai','Kia']
SWD = ['Saab', 'Volvo']
IND = ['Jaguar', 'Land']
USA = ['Buick','Cadillac','Chevrolet','Chrysler','Dodge','Ford','GMC','Hummer' ,'Jeep','Lincoln' ,'Mercury' ,'Oldsmobile' ,'Pontiac' ,'Saturn']

new_column = []
new_continent = []
for col in inp.columns:
  if col == 'Manufacturer':
    for idx in inp.index:
      if inp.loc[idx,col] in GER:
        new_column.append("GER")
        new_continent.append("Europe")
      elif inp.loc[idx,col] in JPN:
        new_column.append("JPN")
        new_continent.append("Asia")
      elif inp.loc[idx,col] in KOR:
        new_column.append("KOR")
        new_continent.append("Asia")
      elif inp.loc[idx,col] in SWD:
        new_column.append("SWD")
        new_continent.append("Europe")
      elif inp.loc[idx,col] in IND:
        new_column.append("IND")
        new_continent.append("Asia")
      elif inp.loc[idx,col] in USA:
        new_column.append("USA") 
        new_continent.append("America")
      elif inp.loc[idx,col] == 'Chrvsler':
        new_column.append("USA") 
        new_continent.append("America")
      elif inp.loc[idx,col] == 'CMC':
        new_column.append("USA")
        new_continent.append("America")
      elif inp.loc[idx,col] == 'Mazda6':
        new_column.append("JPN")
        new_continent.append("Asia")
      else:
        new_column.append("Others")
        print("There is something wrong.. with " + inp.loc[idx,col])             

print(new_column)
print(new_continent)
# inp.insert(1, 'Nation', new_column)
inp.insert(1, 'Continent', new_continent)
```

```python
Drive_Type = inp[inp.columns[10:12]]
Drive_Type['Vehicle Name'] = inp['Vehicle Name']
Drive_Type.head()
```

```python
def Wheel_Type(wheel):
    if wheel['AWD'] == False and wheel['RWD'] == True:
        return "RWD"
    elif wheel['AWD'] == True and wheel['RWD'] == False:
        return "AWD"
    elif wheel['AWD'] == False and wheel['RWD'] == False:
        return "NA"
Drive_Type['Wheel_Drive_Type'] = Drive_Type.apply(Wheel_Type, axis=1)
df5 = Drive_Type[Drive_Type.columns[2:4]]
print(df5)
```

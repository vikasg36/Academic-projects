```python
df = pd.read_csv("input/supermarket_sales .csv")
df.dtypes
```

# New Section

```python
df.isna().sum()
```

```python
df["UNQU"] = df["Unit price"] / df["Quantity"]
df["TATO"] = df["Tax 5%"] / df["Total"]
df["GRTO"] = df["gross margin percentage"] / df["Total"]
df["COTO"] = df["cogs"] / df["Total"]
df["GRGR"] = df["gross margin percentage"] / df["gross income"]

df
```

```python
import altair as alt
from altair import datum
```

```python
alt.Chart(df).mark_circle().encode(
    x='Total:Q',
    y='Tax 5%:Q',
    size='Payment:N',
    color='Product line:N'
).properties(width=1000)
```

```python
line = alt.Chart(df).mark_line().encode(
    alt.X('Unit price:Q',),
    alt.Y('Tax 5%:Q'),
    color="Product line:N"
    ).properties(width=500, height=400)
alt.layer(line)
```

```python
 alt.Chart(df).mark_bar(size=20).encode(
     alt.X("Product line:N"),
     alt.Y("COTO:Q"),
     column="City:N",
 ).properties(width=200)
```

```python
 alt.Chart(df).mark_bar(size=20).encode(
     alt.X("Product line:N"),
     alt.Y("UNQU:Q"),
     column="Payment:N",
 ).properties(width=500)
```

```python
fig = alt.Chart(df).mark_point().encode(x='Unit price',y='Tax 5%')
fig
```

```python
final_plot = fig + fig.transform_regression('Unit price','Tax 5%', method="linear").mark_line(color="#123c1", size=3)
final_plot
```
